```r
source('helpers.R')
source('../utils.R')

df.combined = load_header()
```

```r
df.lot_frontage = df.combined %>% filter(!is.na(LotFrontage))
df.lot_frontage.cutted = df.lot_frontage %>% filter(LotFrontage < 130)
```

```r
options(repr.plot.width = 10, repr.plot.height = 5)

ggplot(df.lot_frontage.cutted, aes(x=BldgType, y=Neighborhood)) +
    geom_tile(aes(fill = LotFrontage)) +
    scale_fill_gradient(low = "green", high = "red") +
    theme_bw()
```

```r
options(repr.plot.width = 10, repr.plot.height = 5)

ggplot(df.lot_frontage.cutted, aes(x=LotArea, y=X1stFlrSF)) +
    geom_raster(aes(fill = LotFrontage)) +
    scale_fill_gradient(low = "green", high = "red") +
    theme_bw()
```
