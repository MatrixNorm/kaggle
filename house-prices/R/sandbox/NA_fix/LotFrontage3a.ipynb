```r
source('helpers.R')
source('../utils.R')

df.combined = load_header()
```

```r
df.lot_frontage = df.combined %>% filter(!is.na(LotFrontage))
df.lot_frontage.cutted = df.lot_frontage %>% filter(LotFrontage < 130)
```

```r
options(repr.plot.width = 10, repr.plot.height = 5)

ggplot(df.lot_frontage.cutted, aes(x=YearBuilt, y=LotFrontage)) +
    geom_count(aes(colour = MSZoning), alpha=0.5) +
    facet_wrap(~MSSubClass) +
    theme_bw()
```

```r
options(repr.plot.width = 10, repr.plot.height = 5)

ggplot(df.lot_frontage.cutted, aes(x=YearBuilt, y=LotFrontage)) +
    geom_count(alpha=0.2) +
    facet_wrap(~LotConfig) +
    theme_bw()
```
