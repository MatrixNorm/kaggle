```r
source('helpers.R')
source('../utils.R')

df.combined = load_header()
```

```r
df.lot_frontage.na = df.combined %>% filter(is.na(LotFrontage))
dim(df.lot_frontage.na)
```

```r
options(repr.plot.width = 10, repr.plot.height = 10)

ggplot(df.lot_frontage.na, aes(x=YearBuilt, y=BldgType)) +
    geom_count(aes(colour = BldgType), alpha=0.5) +
    facet_wrap(~Neighborhood) +
    theme_bw()
```

```r
options(repr.plot.width = 10, repr.plot.height = 5)

df = df.lot_frontage.na %>% filter(BldgType=='1Fam') %>% mutate(LotArea.sqrt=sqrt(LotArea))

ggplot(df, aes(x=YearBuilt, y=LotArea.sqrt)) +
    geom_count(aes(colour = Neighborhood), alpha=0.5) +
    facet_wrap(~Neighborhood) +
    theme_bw()
```
