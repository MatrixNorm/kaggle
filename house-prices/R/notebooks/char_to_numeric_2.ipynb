```r
source('main.R')

data <- kaggle.house$loadData()

df.combined <- kaggle.house$getCombinedDataset(data$train, data$test) %>% (kaggle.house$na$fixAll)

df.training.raw <- 
    df.combined %>% 
    filter(dataSource == "train") %>% 
    mutate(sale_price_log = log(SalePrice)) %>%
    select(-dataSource)
    
df.testing.raw  <- 
    df.combined %>% 
    filter(dataSource == "test") %>% 
    select(-dataSource, -SalePrice)
```

```r
df.training <- kaggle.house$trans$allType1Transform(df.training.raw)
df.testing <- kaggle.house$trans$allType1Transform(df.testing.raw)

test_that("", {
    expect_equal(
        sort(union(colnames(df.training.raw), kaggle.house$trans$tran.column.map %>% unlist %>% as.vector)), 
        sort(colnames(df.training))
    )
})
```

```r
result <- kaggle.house$trans$allType2Transform(list(df = df.training.raw, tran = function(df) { df }))
df.training <- result$df
df.testing <- result$tran(df.testing.raw)

test_that("", {
    
    new.cols <- kaggle.house$trans$type2TransContainer %>% names %>% 
                purrr::map(function (n) { paste0(n, '.new') }) %>% unlist
    
    expect_equal(
        sort(union(colnames(df.training.raw), new.cols)), 
        sort(colnames(df.training))
    )
})

test_that("", {
    
    new.cols <- kaggle.house$trans$type2TransContainer %>% names %>% 
                purrr::map(function (n) { paste0(n, '.new') }) %>% unlist
    
    expect_equal(
        sort(union(colnames(df.testing.raw), new.cols)), 
        sort(colnames(df.testing))
    )
})
```

```r
source('main.R')

res <- kaggle.house$trans$doItAll(df.training.raw, df.testing.raw)

res$df.training %>% colnames %>% sort
res$df.testing %>% colnames %>% sort
```
